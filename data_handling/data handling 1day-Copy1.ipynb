# 데이터결합1

```python
df1 = pd.DataFrame({"A":['A0','A1','A2','A3'],
                    'B':['B0','B1','B2','B3'],
                    'C':['C0','C1','C2','C3'],
                    'D':['D0','D1','D2','D3']},
                  index=[0,1,2,3])

df2 = pd.DataFrame({"A":['A4','A5','A6','A7'],
                    'B':['B4','B5','B6','B7'],
                    'C':['C4','C5','C6','C7'],
                    'D':['D4','D5','D6','D7']},
                  index=[4,5,6,7])

df3 = pd.DataFrame({"A":['A8','A9','A10','A11'],
                    'B':['B8','B9','B10','B11'],
                    'C':['C8','C9','C10','C11'],
                    'D':['D8','D9','D10','D11']},
                  index=[8,9,10,11])

result = pd.concat([df1,df2,df3])
```

```python
result = pd.concat([df1,df2,df3], axis=1)
result
```

```python
result = pd.concat([df1,df2,df3], axis=1, ignore_index = True)
result.columns = ["col_"+str(i) for i in range(12)]
result
```

# 데이터결합2 - 실습

```python
df4 = pd.DataFrame({"A":['A0','A1','A2','A3'],
                    'B':['B0','B1','B2','B3'],
                    'C':['C0','C1','C2','C3'],
                    'D':['D0','D1','D2','D3']})

df5 = pd.DataFrame({"E":['E4','E5','E6','E7'],
                    'F':['F4','F5','F6','F7'],
                    'G':['G4','G5','G6','G7'],
                    'H':['H4','H5','H6','H7']})

df6 = pd.DataFrame({"I":['I8','I9','I10','I11'],
                    'J':['J8','J9','J10','J11'],
                    'K':['K8','K9','K10','K11'],
                    'L':['L8','L9','L10','L11']})

result = pd.concat([df4,df5,df6], axis=1)

result
```

```python
left = pd.DataFrame({"key1":['K0','K1','K2','K3'],
                    'key2':['K0','K1','K0','K1'],
                    'A':['A0','A1','A2','A3'],
                    'B':['B0','B1','B2','B3']})

right = pd.DataFrame({"key1":['K0','K1','K2','K3'],
                    'key2':['K0','K0','K0','K0'],
                    'C':['C0','C1','C2','C3'],
                    'D':['D0','D1','D2','D3']})

result = pd.merge(left,right, on=['key1','key2'])
result
```

```python
result = pd.merge(left,right,on=['key1','key2'], how='left')
result
```

```python
result = pd.merge(left,right,on=['key1','key2'], how='right')
result
```

```python
result = pd.merge(left,right, on=['key1','key2'], how='outer')
result
```

```python
result = pd.merge(left,right, on=['key1','key2'], how='outer', indicator=True)
result
```

# join실습

```python
left = pd.DataFrame({"key1":[1,2,3,4],
                    'A':['A0','A1','A2','A3'],
                    'B':['B0','B1','B2','B3']})

right = pd.DataFrame({'C':['C1','C2','C5'],
                      'D':['D1','D2','D5']},
                    index = [1,2,5])

result = left.join(right)
result
```

```python
result = left.join(right,how='right')
result
```

```python
result = left.join(right, how='inner')
result
```

```python
result = left.join(right, how='outer')
result
```

```python
result = left.join(right, on='key1')
result
```

```python
result2 = pd.merge(left,right, left_on = 'key1', right_index=True, how='left')
result2
```

# pivot실습

```python
df = df[['GENDER','AGEGROUP','WEIGHT']]
new = df.groupby(['GENDER','AGEGROUP']).agg('mean')
new.reset_index(inplace=True)
new.pivot(index='GENDER', columns='AGEGROUP', values='WEIGHT')
```

# stack연습

```python
df = pd.DataFrame({"index":[1,2,3],
                  'alpha':['A','B','C'],
                  'values':[11,22,33],
                  'xyz':['x','y','z']})
df
```

```python
df_stack = df.stack()
print("stack 데이터 크기:" , df_stack.shape)
df_stack
```

```python
df_unstack = df_stack.unstack()
df_unstack
```

# crosstab 연습

```python
df = pd.DataFrame({'factor1':['A','A','A','A','A','B','B','B','B'],
                  'factor2':['a','b','c','a','b','a','b','c','c'],
                  'value':[11,13,12,12,14,12,13,11,12]})
df
```

```python
pd.crosstab(df.factor1, df.factor2)
```

```python
pd.crosstab(df.factor1, df.factor2, rownames=['Fact1'], colnames=['Fact2'],margins=True)
```

```python
pd.crosstab(df.factor1, [df.factor2, df.value])
```

```python
pd.crosstab(df.factor1, df.factor2, margins=True, normalize=True)
```
