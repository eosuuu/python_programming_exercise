```python
# 서포트벡터머신
```

```python
# 데이터구성
```

```python
df = pd.read_csv("유방암.csv", engine='python')
df
```

```python
df_y = df['diagnosis']
df_x = df.drop('diagnosis', axis=1, inplace=False)
df_train_x, df_test_x, df_train_y, df_test_y = train_test_split(df_x, df_y, train_size =0.4, random_state=1234)
df_valid_x, df_test_x, df_valid_y, df_test_y = train_test_split(df_test_x, df_test_y, train_size =0.5, random_state=1234)

print("train data X size : ", df_train_x.shape)
print("train data Y size : ", df_train_y.shape)
print("valid data X size : ", df_valid_x.shape)
print("valid data Y size : ", df_valid_y.shape)
print("test data X size : " ,df_test_x.shape)
print("test data Y size : " ,df_test_x.shape)
```

```python
# step 2 모델 생성
```

```python
# 모델 parameter 조정
```

```python
train_accuracy = []
valid_accuracy = []
para_c = [10**c for c in range(-2,2)]
for v_C in para_c:
    svm = SVC(random_state=1234, C=v_C)
    svm.fit(df_train_x, df_train_y)
    train_accuracy.append(svm.score(df_train_x,df_train_y))
    valid_accuracy.append(svm.score(df_valid_x, df_valid_y))

    
df_accuracy_c = pd.DataFrame()
df_accuracy_c["C"] = para_c
df_accuracy_c["TrainScore"] = train_accuracy
df_accuracy_c["ValidScore"] = valid_accuracy
df_accuracy_c.round(3)
```

```python
plt.plot(para_c, train_accuracy, linestyle="-", label="Train Accuracy")
plt.plot(para_c, valid_accuracy, linestyle="--", label= "Valid Accuracy")
plt.legend()
```

c가 1일때 train값이 과적합을 보이므로 0.1을 선택하였다.

```python
train_accuracy = []
valid_accuracy = []
para_gamma = [10** gamma for gamma in range(-2,2)]
for v_gamma in para_gamma:
    svm = SVC(random_state=1234, C=0.1, gamma=v_gamma)
    svm.fit(df_train_x, df_train_y)
    train_accuracy.append(svm.score(df_train_x,df_train_y))
    valid_accuracy.append(svm.score(df_valid_x, df_valid_y))

    
df_accuracy_gamma = pd.DataFrame()
df_accuracy_gamma["gamma"] = para_gamma
df_accuracy_gamma["TrainScore"] = train_accuracy
df_accuracy_gamma["ValidScore"] = valid_accuracy
df_accuracy_gamma.round(3)
```

```python
plt.plot(para_gamma, train_accuracy, linestyle="-", label="Train Accuracy")
plt.plot(para_gamma, valid_accuracy, linestyle="--", label= "valid_accuracy ")
plt.legend()
```

gamma 변화는 정확도에 영향을 미치지 않는다. 

```python
v_feature_name = df_train_x.columns
scaler = StandardScaler()
df_scaled = scaler.fit_transform(df_x)
df_scaled = pd.DataFrame(df_scaled, columns =v_feature_name)
df_scaled.head()
```

```python
df_scaled_train_x, df_scaled_test_x = train_test_split(df_scaled, train_size =0.4, random_state=1234)
df_scaled_valid_x, df_scaled_test_x = train_test_split(df_scaled_test_x, train_size =0.5, random_state=1234)

print("train data X size : ", df_scaled_train_x.shape)
print("valid data X size : ", df_scaled_valid_x.shape)
print("test data X size : ",df_scaled_test_x.shape)
```

```python
train_accuracy = []
valid_accuracy = []
para_c = [10**c for c in range(-3,3)]
for v_C in para_c:
    svm = SVC(random_state=1234, C=v_C)
    svm.fit(df_scaled_train_x, df_train_y)
    train_accuracy.append(svm.score(df_scaled_train_x,df_train_y))
    valid_accuracy.append(svm.score(df_scaled_valid_x, df_valid_y))

    
df_accuracy_c = pd.DataFrame()
df_accuracy_c["C"] = para_c
df_accuracy_c["TrainScore"] = train_accuracy
df_accuracy_c["ValidScore"] = valid_accuracy
df_accuracy_c.round(3)
```

```python
plt.plot(para_c, train_accuracy, linestyle="-", label="Train Accuracy")
plt.plot(para_c, valid_accuracy, linestyle="--", label= "Valid Accuracy")
plt.legend()
```

C는 10일때 가장 모델의 정확도가 많이증가하므로 10으로 선택했다. 

```python
train_accuracy = []
valid_accuracy = []
para_gamma = [10** gamma for gamma in range(-2,2)]
for v_gamma in para_gamma:
    svm = SVC(random_state=1234, C=10, gamma=v_gamma)
    svm.fit(df_scaled_train_x, df_train_y)
    train_accuracy.append(svm.score(df_scaled_train_x,df_train_y))
    valid_accuracy.append(svm.score(df_scaled_valid_x, df_valid_y))

    
df_accuracy_gamma = pd.DataFrame()
df_accuracy_gamma["gamma"] = para_gamma
df_accuracy_gamma["TrainScore"] = train_accuracy
df_accuracy_gamma["ValidScore"] = valid_accuracy
df_accuracy_gamma.round(3)
```

```python
plt.plot(para_gamma, train_accuracy, linestyle="-", label="Train Accuracy")
plt.plot(para_gamma, valid_accuracy, linestyle="--", label= "Valid Accuracy")
plt.legend()
```

gamma는 0.01일때 모델의 정확도가 가장 높으므로 0.01을 선택하였다.

```python
svm_final = SVC(random_state=1234, C=10, gamma=0.01)
svm_final.fit(df_scaled_train_x, df_train_y)
y_pred = svm_final.predict(df_scaled_test_x)

print("Accuracy on training set :", svm_final.score(df_scaled_train_x, df_train_y))
print("Accuracy on test set :", svm_final.score(df_scaled_test_x, df_test_y))
print("Confusion matrix:\n", confusion_matrix(df_test_y, y_pred))

```

training set의 정확도는 99.55%이고, test set은 95.90%인 것으로 산출되었다.   
confusion matrix를 보면 TN과 TP가 102와 62인것을 알 수 있다.   
특히 FN이 7이고, FP가 0인것을 살펴보면 유방암인데 유방암이 아니라고 진단할 확률이 유방암이 아닌데 유방암이라고 진단할 확률보다 높다.  
7이라는 수치는 적은 수치라고 볼 수 있지만 의료산업에서는 1%의 가능성도 큰 확률이라고 생각하기 때문에 FP보다 FN이 0에 가까워야한다고 생각한다.  
의료 모델에 산출하기 위해서 FP의 수치가 증가하는 한이 있어도 FN의 수치를 줄이기 위해 더 노력해야겠다.
