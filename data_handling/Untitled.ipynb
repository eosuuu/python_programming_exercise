# 연령대 column 삽입

```python
def divide_age(age):
    if age >= 0 and age<20 : 
        result = "10대"
    elif age>=20 and age<30:
        result = "20대"
    elif age>=30 and age<40:
        result = "30대"
    elif age>=40 and age<50:
        result = "40대"
    elif age>=50 and age<60:
        result = "50대"
    elif age>=60 and age<70:
        result = "60대"
    elif age>=70 and age<80:
        result = "70대"
    else:
        result = "80대"
    return result
```

```python
def divide_position(name):
    if 'Mr' in name:
        result = "결혼남"
    elif 'Mrs' in name:
        result = '결혼여'
    elif 'Miss' in name:
        result = '결혼X'
    else :
        result = '결혼X'
    
    return result
```

```python
def divide_cabin(cabin):
    if 'A' in cabin:
        result = "A그룹"
    elif 'B' in cabin:
        result = "B그룹"
    elif 'C' in cabin:
        result = "C그룹"
    else :
        result = "기록X"
    return result
```

```python
df = pd.read_csv("titanic.csv",engine='python')
agegroup = []
name = []
cabin = []

for i in df['Age']:
    agegroup.append(divide_age(i))
for a in df["Name"]:
    name.append(divide_position(a))
for c in df['Cabin']:
#     cabin.append(divide_cabin(c))
    print(c)
    
    
    
df['AgeGroup'] = agegroup
df['CabinGroup'] = cabin
# df['Name'] = name
# df.sort_index('cabin')
```
