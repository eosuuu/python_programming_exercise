```python
import pandas as pd
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import seaborn as sns

# matplotlib.rc('font',family='Malgun Gothic')
```

```python
fit = pd.read_csv("FITNESS.csv", engine='python', encoding='CP949')
cur = pd.read_csv("환율.csv",engine='python', encoding="CP949",parse_dates=['APPL_DATE'])
fit
```

```python
gen_count = fit['GENDER'].value_counts()
gen_count.plot.barh()
```

```python
age_count = fit['AGEGROUP'].value_counts()
age_count.sort_index(inplace = True)
age_count.plot.barh()
```

```python
group_count = fit.groupby(['GENDER', 'AGEGROUP'], as_index = False).size()
group_count.plot.barh(color=['red','green','blue','cyan','magenta','black'])
```

```python
cur_pivot['CNY'].plot()
```

```python
cur_pivot[['JPY','USD','CNY']].plot()
```

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import scale, minmax_scale, robust_scale
%matplotlib inline
cur_pivot_std = scale(cur_pivot)
cur_pivot_std = pd.DataFrame(cur_pivot_std, columns=cur_pivot.columns)
cur_pivot_std[['CNY','JPY','USD']].plot()
```

```python
agg = fit.groupby(['GENDER','AGEGROUP'],as_index=False).agg('mean')
agg_pivot = agg.pivot(index="GENDER", columns="AGEGROUP", values='OXY')
sns.heatmap(agg_pivot, cmap='Blues')
```

```python
sns.kdeplot(fit['RUNTIME'], fit['OXY'],shade=False,cbar=True)
sns.scatterplot(x='RUNTIME', y="OXY", data = fit)
plt.show()
```

```python
sns.kdeplot(fit['RUNTIME'], fit['OXY'],shade=True,cbar=True, n_levels=10)
sns.scatterplot(x='RUNTIME', y="OXY", data = fit)
plt.show()
```

```python
pd.plotting.parallel_coordinates(fit,'GENDER', cols=['OXY', 'WEIGHT', 'RSTPULSE'],colormap='Accent')
```

```python
df_count = fit["AGEGROUP"].value_counts()
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(13,10))
plt.tight_layout(w_pad=5, h_pad=5)

axes[0,0].hist(fit["OXY"])
axes[0,0].set_title("Histogram", fontsize=15)
axes[0,0].set_xlabel("혈당 산소 요구량", fontsize=12)

axes[0,1].pie(df_count, labels=df_count.index.tolist(), autopct="%.1f%%")
axes[0,1].set_title("Pie Chart", fontsize=15)
axes[0,1].set_xlabel("연령대", fontsize=12)

axes[1,0].plot("RUNTIME", "RUNPULSE", data=fit, label="맥박(운동)")
axes[1,0].plot("RUNTIME", "OXY", data=fit, label="혈당 산소 요구량")
axes[1,0].set_title("Trend", fontsize=15)
axes[1,0].set_xlabel("운동시간", fontsize=12)
axes[1,0].legend()

axes[1,1].scatter("RUNPULSE", "OXY", data=fit)
axes[1,1].set_title("Scatter", fontsize=15)
axes[1,1].set_xlabel("맥박(운동)", fontsize=12)
axes[1,1].set_ylabel("혈당 산소 요구량", fontsize=12)

```
