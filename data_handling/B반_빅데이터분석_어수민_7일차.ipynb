# 데이터구성

```python
df = pd.read_csv("유방암.csv", engine='python')
df
```

```python
df_y = df['diagnosis']
df_x = df.drop('diagnosis', axis=1, inplace=False)
df_train_x, df_test_x, df_train_y, df_test_y = train_test_split(df_x, df_y, train_size =0.4, random_state=1234)
df_valid_x, df_test_x, df_valid_y, df_test_y = train_test_split(df_test_x, df_test_y, train_size =0.5, random_state=1234)

print("train data X size : ", df_train_x.shape)
print("train data Y size : ", df_train_y.shape)
print("valid data X size : ", df_valid_x.shape)
print("valid data Y size : ", df_valid_y.shape)
print("test data X size : " ,df_test_x.shape)
print("test data Y size : " ,df_test_x.shape)
```

# step 2 모델 생성

# 모델 parameter 조정

```python
train_accuracy = []
valid_accuracy = []
para_c = [10**c for c in range(-2,2)]
for v_C in para_c:
    svm = SVC(random_state=1234, C=v_C)
    svm.fit(df_train_x, df_train_y)
    train_accuracy.append(svm.score(df_train_x,df_train_y))
    valid_accuracy.append(svm.score(df_valid_x, df_valid_y))

    
df_accuracy_c = pd.DataFrame()
df_accuracy_c["C"] = para_c
df_accuracy_c["TrainScore"] = train_accuracy
df_accuracy_c["ValidScore"] = valid_accuracy
df_accuracy_c.round(3)
```

```python
plt.plot(para_c, train_accuracy, linestyle="-", label="Train Accuracy")
plt.plot(para_c, valid_accuracy, linestyle="--", label= "Valid Accuracy")
plt.legend()
```

c가 1일때 train값이 과적합을 보이므로 0.1을 선택하였다.

```python
train_accuracy = []
valid_accuracy = []
para_gamma = [10** gamma for gamma in range(-2,2)]
for v_gamma in para_gamma:
    svm = SVC(random_state=1234, C=0.1, gamma=v_gamma)
    svm.fit(df_train_x, df_train_y)
    train_accuracy.append(svm.score(df_train_x,df_train_y))
    valid_accuracy.append(svm.score(df_valid_x, df_valid_y))

    
df_accuracy_gamma = pd.DataFrame()
df_accuracy_gamma["gamma"] = para_gamma
df_accuracy_gamma["TrainScore"] = train_accuracy
df_accuracy_gamma["ValidScore"] = valid_accuracy
df_accuracy_gamma.round(3)
```

```python
plt.plot(para_gamma, train_accuracy, linestyle="-", label="Train Accuracy")
plt.plot(para_gamma, valid_accuracy, linestyle="--", label= "valid_accuracy ")
plt.legend()
```

gamma 변화는 정확도에 영향을 미치지 않는다.

# knn분석

```python
df = pd.read_csv("유방암.csv", engine='python')
df
```

```python
df_y = df['diagnosis']
df_x = df.drop('diagnosis', axis=1, inplace=False)
df_train_x, df_test_x, df_train_y, df_test_y = train_test_split(df_x, df_y, train_size =0.4, random_state=1234)
df_valid_x, df_test_x, df_valid_y, df_test_y = train_test_split(df_test_x, df_test_y, train_size =0.5, random_state=1234)

print("train data X size : ", df_train_x.shape)
print("train data Y size : ", df_train_y.shape)
print("valid data X size : ", df_valid_x.shape)
print("valid data Y size : ", df_valid_y.shape)
print("test data X size : " ,df_test_x.shape)
print("test data Y size : " ,df_test_x.shape)
```

# step 2 모델 생성

# 모델 parameter 조정

```python
train_accuracy = []
valid_accuracy = []
para_n_neighbors = [i for i in range(1,16)]
for v_n_neighbors in para_n_neighbors:
    knn = KNeighborsClassifier(n_neighbors=v_n_neighbors)
    knn.fit(df_train_x, df_train_y)
    train_accuracy.append(knn.score(df_train_x,df_train_y))
    valid_accuracy.append(knn.score(df_valid_x, df_valid_y))

    
df_accuracy_neighbors = pd.DataFrame()
df_accuracy_neighbors["Neighbors"] = para_n_neighbors
df_accuracy_neighbors["TrainScore"] = train_accuracy
df_accuracy_neighbors["ValidScore"] = valid_accuracy
df_accuracy_neighbors.round(3)
```

```python
plt.plot(para_n_neighbors, train_accuracy, linestyle="-", label="Train Accuracy")
plt.plot(para_n_neighbors, valid_accuracy, linestyle="--", label= "Valid Accuracy")
plt.legend()
```

neighbors의 수는 5일때 가장 모델의 정확도가 높으므로 5를 선택하였다.

```python
train_accuracy = []
valid_accuracy = []
para_n_neighbors = [i for i in range(1,11)] * 2
para_weights = (['uniform']) * 10 + (['distance']*10)

for(v_n_neighbors, v_weights) in zip(para_n_neighbors, para_weights):
    knn = KNeighborsClassifier(n_neighbors=v_n_neighbors, weights=v_weights)
    knn.fit(df_train_x, df_train_y)
    train_accuracy.append(knn.score(df_train_x,df_train_y))
    valid_accuracy.append(knn.score(df_valid_x,df_valid_y))
    
df_accuracy_weights = pd.DataFrame()
df_accuracy_weights['Neighbors']=para_n_neighbors
df_accuracy_weights['Weights']=para_weights
df_accuracy_weights['TrainAccuracy']=train_accuracy
df_accuracy_weights['ValidAccuracy']=valid_accuracy
df_accuracy_weights.round(3)

```

```python
df_accuracy_weights_pivot = df_accuracy_weights.pivot(index = 'Neighbors', columns='Weights',
                                                     values=['TrainAccuracy', 'ValidAccuracy'])
df_accuracy_weights_pivot
```

```python
sns.lineplot(data=df_accuracy_weights_pivot)
```

```python
level0 = df_accuracy_weights_pivot.columns.get_level_values(0)
level1 = df_accuracy_weights_pivot.columns.get_level_values(1)
df_accuracy_weights_pivot.columns = level0 + "_" + level1
df_accuracy_weights_pivot.head()
```

```python
sns.lineplot(data=df_accuracy_weights_pivot)
```

neighbors가 5일때 distance보다 uniform이 더 모델의 정확도가 높으므로, uniform을 선택하였다.

```python
train_accuracy = []
valid_accuracy = []
para_n_neighbors = [i for i in range(1,11)] * 3
para_metric = ['minkowski'] * 10 + ['euclidean']*10 + ['manhattan']*10


for(v_n_neighbors, v_metric) in zip(para_n_neighbors, para_metric):
    knn = KNeighborsClassifier(n_neighbors=v_n_neighbors, weights='uniform', metric=v_metric)
    knn.fit(df_train_x, df_train_y)
    train_accuracy.append(knn.score(df_train_x,df_train_y))
    valid_accuracy.append(knn.score(df_valid_x,df_valid_y))
    
df_accuracy_metric = pd.DataFrame()
df_accuracy_metric['Neighbors']=para_n_neighbors
df_accuracy_metric['Metric']=para_metric
df_accuracy_metric['TrainAccuracy']=train_accuracy
df_accuracy_metric['ValidAccuracy']=valid_accuracy
df_accuracy_metric.round(3)

```

```python
df_accuracy_metric_pivot = df_accuracy_metric.pivot(index = 'Neighbors', columns='Metric',
                                                     values=['TrainAccuracy', 'ValidAccuracy'])
level0 = df_accuracy_metric_pivot.columns.get_level_values(0)
level1 = df_accuracy_metric_pivot.columns.get_level_values(1)
df_accuracy_metric_pivot.columns = level0 + "_" + level1
sns.lineplot(data=df_accuracy_metric_pivot)
```

neighbors가 7일때, 거리계산은 manhattan일때 가장 모델의 정확도가 높은 것을 알 수있다. 

```python
knn_final = KNeighborsClassifier(n_neighbors=7, weights='uniform', metric='manhattan')
knn_final.fit(df_train_x, df_train_y)
y_pred = knn_final.predict(df_test_x)

print("Accuracy on training set :", knn_final.score(df_train_x, df_train_y))
print("Accuracy on test set :", knn_final.score(df_test_x, df_test_y))
print("Confusion matrix:\n", confusion_matrix(df_test_y, y_pred))

```

서포트벡터머신의 training set의 정확도는 99.55%이고, test set은 95.90%이었다.  
knn의 training set의 정확도는 95.15%이고, test set은 92.39%이다.   
confusion matrix를 보면 TN과 TP가 98과 60인것을 알 수 있다.   
특히 FN이 9이고, FP가 4인 것을 보면 어느정도 오류가 검출되는 것을 알수있다.   
서포트벡터머신의 FN은 7인것과 비교하면 knn보단 서포트벡터머신이 모델의 정확도도 높은 것을 알 수있다.

인공신경망 모델의 confusion matrix를 보면 서포트벡터머신의 matrix와 큰 차이가 없는 것을 볼 수 있다.   
하지만 FN값이 1감소한 6이기 때문에 의료모델에 적용한다면 세가지 모델중에 인공신경망 모델이 가장 적합한것으로 판단된다.   
모델의 정확도는 95%이상으로 높은편이지만, 의료업계라는 특수한 상황에 비추어 봤을때는 적어도 99%는 되어야할 것으로 판단되고,   
FN값이 더 감소해야 의미있는 모델이라고 볼 수 있다.
