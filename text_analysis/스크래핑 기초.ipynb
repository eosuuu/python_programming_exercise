```python
# beautiful soup 라이브러리 사용
```

```python
from bs4 import BeautifulSoup
```

```python
html = """
<ul>
    <li>1교시</li>
    <li>2교시</li>
</ul>

<ol>
    <li>1교시</li>
    <li>1교시</li>
</ol>
"""

soup = BeautifulSoup(html, 'html5lib')
temp = soup.select("ul li")

for word in temp:
    print(word.text)
```

```python
html = ''' 
<table>
    <tr>
        <th> 재무정보 </th>
        <th> 2018 </th>
        <th> 2019 </th>
    </tr>
    <tr>
        <td> 매출액 </td>
        <td> 100 </td>
        <td> 200 </td>
    </tr>
</table>
'''

soup = BeautifulSoup(html, 'html5lib')
result = soup.select('td')

for a in result:
    print(a.text,end='/')
    
    
```
