```python
t = 1.53
df = 5

prob = stats.t.cdf(t,df)

print(prob)
```

```python
chi2 = 10
df = 30

prob = stats.chi2.cdf(chi2, df)

print(prob)
```

```python
prob2 = stats.chi2.cdf(3.0, 8)

print(prob2)
```

```python
f = 1.5
df1 = 10
df2 = 100
prob = stats.f.cdf(f,df1,df2)

print(prob)
```

```python
x = 1500
a = 2.2
b = 1200

prob = stats.weibull_min.cdf(x,a,scale =b)

print(1-prob)
```

```python
x = 2500
a = 5.2
b = 2200
prob = stats.weibull_min.cdf(x,a,scale=b)

print(1-prob)
```

```python
n = 3 
result = 0
for i in range(n+1):
    prob = stats.binom.pmf(k=i,n=n,p=0.4)
    print(prob)
    result += prob
    
print(result)
```

```python
n = 5
result = 0
for i in range(n+1):
    prob = stats.binom.pmf(k=i, n=n, p=0.3)
    print(prob)
    result += prob

print(result)
```

```python
mu = 2 

pmf_prob = stats.poisson.pmf(3,mu)
cdf_prob = stats.poisson.cdf(2,mu)

print(pmf_prob)
print(cdf_prob)
```

```python
e = 3
a = stats.poisson.pmf(2,e)
b = stats.poisson.cdf(1,e)
c = stats.poisson.cdf(2,e)
print(a, b, 1-c)
```
