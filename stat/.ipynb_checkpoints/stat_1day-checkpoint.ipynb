```python
mu = 115
sigma = 20
x= 135
prob = stats.norm.cdf(x,mu,sigma)
print("{0}   A/S   : {1:.1f}%".format(x,(1-prob)*100))
print("%s   A/S   : %.1f" %(x,(1-prob)*100)+"%")
```
