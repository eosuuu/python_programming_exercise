```python
mu = 115
sigma = 20
x= 135
prob = stats.norm.cdf(x,mu,sigma)
print("{0}분 이상 걸리는 A/S 작업 비율 : {1:.1f}%".format(x,(1-prob)*100))
print("%s분 이상 걸리는 A/S 작업 비율 : %.1f" %(x,(1-prob)*100)+"%")
```

```python
mu = 175
sigma = 5
x = 180
prob = stats.norm.cdf(x,mu,sigma)
print("%s이상 걸리는 비율 : %.1f" %(x,(1-prob)*100)+"%")
```

```python
mu = 12
sigma = 3
x = 15
prob = stats.norm.cdf(x,mu,sigma)
print("%dkm 이상 %dkm 이하를 달릴 확률은 %.1f" %(mu,x,(prob - 0.5)*100)+"%")
```
